```python
import numpy
import pandas
```

```python
# Dataframe creation
data = pandas.read_csv('heart (1).csv')
df = pandas.DataFrame(data)
df
```

```python
# Target  >>>  HeartDisease
```

```python
# Unnecessory column removal
df = df.drop(columns= ['Age','Sex'],axis=1)
df
```

```python
# Checking missing values
df.isna().sum()
```

```python
# Encoding

# Class = Label Encoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['ChestPainType'] = le.fit_transform(df['ChestPainType'])
df['RestingECG'] = le.fit_transform(df['RestingECG'])
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
df['ST_Slope'] = le.fit_transform(df['ST_Slope'])
df
```

```python
X = df.iloc[:,:-1]
X
Y = df.iloc[:,-1]
Y
```

Data preprocessing steps
=====================================
Handling Missing Values
Data Transform
Outliers
Scaling
Data Reduction

```python
# Splitting
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)

X_train
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler    # Z = (x - mean) / Standard_Deviation

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)     # 70 %      fit takes mean,std of 70% data

X_test = scaler.transform(X_test)   # 30 %%!    X_test is transformed using mean,std,etc.. of the 70 % data
                                               #Hence model does not see the 30 % data
X_train
X_test
```

```python
# Algorithm
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
```

```python
# Model Creation
model = knn.fit(X_train,Y_train)
```

```python
# Testing
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_pred)
cm
```
