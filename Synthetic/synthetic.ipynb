```python
import numpy
import pandas
```

```python
data = pandas.read_csv("synthetic_dataset.csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
#result >> stock
# input >>> Category, price, discount
# removing column Category as the missing values of the column exceeds 50%
```

```python
df = df.drop('Category',axis=1)
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset = 'Stock')    # removing rows where stock value is missing
df
```

```python
df.isna().sum()
```

```python
df['Price'] = df['Price'].fillna(df['Price'].mean())    # filling missing values of price column with mean of the column
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())   # filling missing values of rating column with mean of the column
```

```python
df['Discount'] = df['Discount'].fillna(df['Discount'].mean())   # filling missing values of discount column with mean of the column
```

```python
df.isna().sum()
```

```python
df
```

Converting values in Column Stock to numeric values
This process is called Encoding

```python
df['Stock'] = df['Stock'].replace({'In Stock' : 1, 'Out of Stock' : 0})
```

```python
df.isna().sum()
```

```python
X = df.loc[:, df.columns != 'Stock']
X
Y = df['Stock']
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

model = knn.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_pred)
cm
```
