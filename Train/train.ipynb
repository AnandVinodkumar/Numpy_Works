```python
import numpy
import pandas
```

```python
data = pandas.read_csv("train.csv")
df = pandas.DataFrame(data)
df
```

```python
#1)
df.head(10)
#2)
df.shape
#3)
df.dtypes
#4) Convert order date and ship date to datetime
df['Order Date'] = pandas.to_datetime(df['Order Date'],format='%d/%m/%Y')
df['Ship Date'] = pandas.to_datetime(df['Ship Date'],format='%d/%m/%Y')

#5)
df.isna().sum()
#6) Remove all duplicate rows
df.drop_duplicates(inplace=True)
df.shape
#7)
df.rename(columns={'Row ID':'row_id','Order ID':'order_id'}, inplace=True)
df
#8)
df.sort_values(by='Sales', ascending=False)
#9)
df[df['Sales'] > 500]
#10)
df[df['Ship Mode'] == 'Second Class']
#11)
df[df['State'] == 'California']
#12) 
df[(df['Segment'] == 'Consumer') & (df['Sales'] > 200)]
#13)
df[(df['Category'] == 'Furniture') & (df['Region'] == 'West')]
#14)
df[df['Ship Date'] > df['Order Date']]
#15) Create a new column 'Order_Year' from 'Order Date'
df['Order_Year'] = df['Order Date'].dt.year
df
#16)
df['Shipping_Time_days'] = df.apply(lambda a: (a['Ship Date'] - a['Order Date']).days, axis=1)
#17)
df['Profit_Category'] = df.apply(lambda a: "High" if a['Sales'] > 500 else ("Medium" if (a['Sales'] >= 100) & (a['Sales'] <= 500) else "Low"),axis=1)
df
#18)
df['First_Name'] = df.apply(lambda a: a['Customer Name'].split(" ")[0], axis=1)
df
#19)
df['discount_amt'] = df.apply(lambda a: a['Sales'] * 0.1, axis=1)
df
#20)
df['City'] = df['City'].str.lower()
#21)
df['Product Name'] = df['Product Name'].replace({' ': '_'})
df
```
