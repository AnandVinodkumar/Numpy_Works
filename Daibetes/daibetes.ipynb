```python
import numpy
import pandas
```

```python
data = pandas.read_csv("diabetes.csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
# target >>> outcome
# input features >>> Pregnancies	Glucose	BloodPressure	SkinThickness	Insulin	BMI	DiabetesPedigreeFunction	Age

# No missing values so no need of filling

# scikit-learn >>> Contains tools to build, train, test and evaluating ML models easily
# ==========================================================================================================
```

```python
# Splitting
#=========================
X = df.iloc[:,:-1]

Y = df.iloc[:,-1]
Y
```

Input Features >>> 70 % training  30 % testing >>> X_train  X_test
Target >>>         70 % training  30 % evaluation >>> Y_train  Y_test

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)

X_train
```

```python
# Z = (x - mean) / standardDeviation    >>>   Standard Scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_train

"""
No need to give X_test to fit method
"""
X_test = scaler.transform(X_test)
X_test
```

```python
"""
KNN algorithm 
"""
# Training
from sklearn.neighbors import KNeighborsClassifier  # class method KNN
knn = KNeighborsClassifier()
knn = knn.fit(X_train,Y_train)
```

```python
# Testing
y_pred = knn.predict(X_test)    # Model will predict the result for untrained data
y_pred
```

PERFORMANCE EVALUATION

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,Y_test))
```

```python
# Acccuracy Score = correct predictions / total no. of predictions
# Recall
# Precision
# f1-score
# Confusion matrix (Only for classification)
```

```python
# Confusion matrix
# TruePositive, TrueNegative, FalsePositive, FalseNegative

# real result      model result
#     1                 1       #True Positive
#     1                 0       #False Negative
#     0                 1       #False Positive
#     0                 0       #True Negative
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_pred)
cm
```
