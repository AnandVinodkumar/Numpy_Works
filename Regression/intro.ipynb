Simple Linear Regression
======================================

"In simple linear regression, it ca predict a result using one input feature"

area          House_price (Dependent)

2600          5600000

2800          5800000

3100          6100000

3400          6600000

3800          7200000


Testing   area = 3600   price = ???

```python
area = [2600,2800,3100,3400,3800]

price = [5600000, 5800000, 6100000, 6600000, 7200000]
```

```python
from matplotlib import pyplot

pyplot.scatter(area, price)
pyplot.xlabel("Area")
pyplot.ylabel("Price")
pyplot.title("Area vs Price")
pyplot.show()
```

```python
# y = mx + c

# y = output

# m = slope

# x = independent variable

# c = y intercept
```
