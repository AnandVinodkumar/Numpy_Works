Simple Linear Regression
======================================

"In simple linear regression, it can predict a result using one input feature"

area          House_price (Dependent)

Testing   area = 3600   price = ???

```python
area = [2600,2800,3100,3400,3800]  # independent variable

price = [500000, 600000, 700000, 800000, 900000]  # dependent variable
```

```python
from matplotlib import pyplot

pyplot.scatter(area, price)
pyplot.xlabel("Area")
pyplot.ylabel("Price")
pyplot.title("Area vs Price")
pyplot.show()
```

```python
# y = mx + c

# y = output

# m = slope

# x = independent variable

# c = y intercept
```

```python
hours = [1,2,3,4]
marks = [40,50,60,70]
pyplot.scatter(hours, marks)
pyplot.xlabel("Hours Studied")
pyplot.ylabel("Marks Obtained")
pyplot.title("Hours Studied vs Marks Obtained")
pyplot.show()
```
