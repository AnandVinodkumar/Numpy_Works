```python
import pandas
```

```python
data = pandas.read_csv("poly.csv")
df = pandas.DataFrame(data)
df
```

```python
X = df.iloc[:,1:2]
Y = df.iloc[:,-1]
```

```python
from matplotlib import pyplot

pyplot.scatter(X,Y,color="red")
pyplot.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

model = lr.fit(X,Y)

Y_pred = model.predict(X)

Y_pred
```

```python
Y
```

```python
from sklearn.preprocessing import PolynomialFeatures

feature = PolynomialFeatures(degree=3)

X_poly = feature.fit_transform(X)
```

```python
model1 = lr.fit(X_poly,Y)
Y_poly = model1.predict(X_poly)
Y_poly
```

```python
Y
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score

print(mean_absolute_percentage_error(Y,Y_poly))
print(r2_score(Y,Y_poly))
```
