```python
import numpy
import pandas
```

```python
data = pandas.read_csv("taxi_trip_pricing (1).csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Trip_Price'])
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
for i in df.columns:

    if df[i].dtype == 'object':

        df[i] = df[i].fillna(df[i].mode()[0])
    
    else:

        df[i] = df[i].fillna(df[i].mean())
```

```python
df.isna().sum()
```

```python
df
```

```python
"""
One-Hot Encoder
"""
dfe = pandas.get_dummies(df[['Time_of_Day','Day_of_Week','Traffic_Conditions','Weather']],dtype=int)
dfe
```

```python
# Combine

df_final = pandas.concat([df,dfe],axis=1)
df_final
```

```python
df_final = df_final.drop(['Time_of_Day','Day_of_Week','Traffic_Conditions','Weather'],axis=1)
df_final
```

```python
X = df_final.drop(['Trip_Price'],axis=1)
Y = df_final['Trip_Price']
```

```python
# """
# One-Hot Encoder
# """
# X = pandas.get_dummies(X,drop_first=True,dtype=int)
# X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

model = lr.fit(X_train,Y_train)

Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score

print(mean_absolute_percentage_error(Y_test,Y_pred))
print(r2_score(Y_test,Y_pred))
```
