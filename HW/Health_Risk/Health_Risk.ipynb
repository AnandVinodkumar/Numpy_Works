```python
import numpy
import pandas
```

```python
data = pandas.read_csv("Lifestyle_and_Health_Risk_Prediction_Synthetic_Dataset.csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df['exercise'] = df['exercise'].replace({'none' : 0, 'low' : 1, 'medium' : 2, 'high' : 3})
```

```python
df['sugar_intake'] = df['sugar_intake'].replace({'low': 0,'medium':1,'high':2})
```

```python
df['smoking'] = df['smoking'].replace({'yes' : 1, 'no' : 0})
```

```python
df['alcohol'] = df['alcohol'].replace({'yes' : 1, 'no' : 0})
```

```python
df['married'] = df['married'].replace({'yes' : 1, 'no' : 0})
```

```python
# df['profession'].unique()    >>>  To find what all values are there in the column
df['profession'] = df['profession'].replace({'office_worker' : 0, 'teacher' : 1, 'artist' : 2, 'farmer' : 3, 'driver' : 4, 'engineer' : 5, 'student' : 6, 'doctor' : 7})
```

```python
df['health_risk'] = df['health_risk'].replace({'low' : 0, 'high' : 1})
```

```python
df
```

```python
# input features >>> age	weight	height	exercise	sleep	sugar_intake	smoking	alcohol	married	profession	bmi
# output    >>>    health_risk
```

```python
X = df.iloc[:,:-1]
X
Y = df.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)

X_train
X_test
Y_train
Y_test
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)

X_train

X_test = scaler.transform(X_test)
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn = knn.fit(X_train,Y_train)
```

```python
Y_pred = knn.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_pred,Y_test))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,Y_pred)
cm
```
