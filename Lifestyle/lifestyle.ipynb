```python
import numpy
import pandas
```

```python
data = pandas.read_csv("Lifestyle_and_Health_Risk_Prediction_Synthetic_Dataset (1).csv")
df = pandas.DataFrame(data)
df
```

```python
df = df.drop(columns=['age','married'])
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['exercise'] = le.fit_transform(df['exercise'])
df['sugar_intake'] = le.fit_transform(df['sugar_intake'])
df['smoking'] = le.fit_transform(df['smoking'])
df['alcohol'] = le.fit_transform(df['alcohol'])
df['profession'] = le.fit_transform(df['profession'])
df['health_risk'] = le.fit_transform(df['health_risk'])
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

model = knn.fit(X_train,Y_train)

Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test,Y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,Y_pred)

cm
```
