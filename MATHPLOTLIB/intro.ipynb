```python
from matplotlib import pyplot
```

```python
year = [2010,2011,2012,2013,2014]

salary = [1000,1500,2000,2500,3000]

pyplot.plot(year,salary)

pyplot.show()
```

```python
height = [150,153,147,157,155]

weight = [80,87,67,72,77]

pyplot.plot(height,weight)

pyplot.show()
```

```python
x = [2,4,6,8,10]    # Both x and y have the same intervals

y = [3,6,9,12,15]   # Hence the graph will be a straight line

pyplot.plot(x,y)

pyplot.show()
```

```python
x = [2,4,6,8,10] 

y = [3,6,9,12,15]   

pyplot.scatter(x,y)

pyplot.show()
```

```python
height = [146,150,152,153,140]

weight = [80,85,90,88,95]

pyplot.scatter(height,weight,color = "red")

pyplot.xlabel("Height in cm",fontsize = 15)

pyplot.ylabel("Weight in kg")

pyplot.title("Height-Weight graph")

pyplot.show()
```

```python
x = [2,4,6,8,10]

y = [3,6,9,12,15]

x1 = [1,5,10,15,20]

y1 = [20,23,26,29,32]

pyplot.plot(x,y)

pyplot.plot(x1,y1)

pyplot.show()
```

```python
# Bar chart

x = [2,4,6,8,10]

y = [3,6,9,12,15]

pyplot.bar(x,y)

pyplot.show()
```

```python
# Histogram
#==================================
# Used to represent the no of frequency in a certain range
#===========================================================================

marks = [10,15,20,21,25,30,31,32,33,35,40]

pyplot.hist(marks,edgecolor = "black")

pyplot.show()
```

PIE CHART

```python
activities = ['sleeping','travel','work','other']

hours = [8,4,8,4]

colors = ['red','green','skyblue','cyan']

explode = (0.1,0,0,0)

pyplot.pie(hours,labels=activities,colors=colors,explode=explode,autopct='%1.1f%%',startangle=90,shadow=True)

pyplot.show()
```
