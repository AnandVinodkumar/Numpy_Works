```python
import seaborn

import pandas

from matplotlib import pyplot
```

```python
days = ["mon","tue","wed","thurs","fri","sat"]

sales = [120,150,160,190,200,195]

data = pandas.DataFrame({"days":days,"sales":sales})

data

```

```python
seaborn.lineplot(data=data,x="days",y="sales",marker="o")

pyplot.title("Sales of Week")
pyplot.show()
```

```python
employee_data = {
    "depts" : ["IT","HR","Testing","Sales"],
    "salaries" : [30000,40000,35000,50000]
}

employee_df = pandas.DataFrame(employee_data)
employee_df
```

```python
seaborn.boxplot(data=employee_df,x="depts",y="salaries")

pyplot.title("Employee data")
pyplot.show()
```

```python
ages = [20,21,21,22,20,30,32,32,30,33,23,23,23]

seaborn.histplot(data=ages)

pyplot.title("Employee Age")
pyplot.xlabel("Ages")
pyplot.show()
```

```python
employee_data = {
    "depts" : ["IT","HR","Testing","Sales"],
    "salaries" : [30000,40000,35000,50000]
}

# Bargraph

df = pandas.DataFrame(employee_data)

seaborn.barplot(data=df,x="depts",y="salaries",palette="coolwarm",hue="salaries")

pyplot.show()
```

```python
# Create a barchart using two lists

subjects = ["chem","phy","maths","biology"]
marks = [80,90,100,95]

df = pandas.DataFrame({"Subjects":subjects,"Marks":marks})

seaborn.barplot(data=df,x="Subjects",y="Marks",palette="coolwarm",hue="Marks")
pyplot.show()
```
