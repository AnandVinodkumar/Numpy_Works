```python
import numpy
import pandas
from matplotlib import pyplot
```

```python
data = pandas.read_csv("sales_data_clean.csv")
df = pandas.DataFrame(data)
df
```

```python
df.dtypes
```

```python
product_sales = [df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
items = ['facecream','facewash','toothpaste','bathsoap','shampoo','moisturizer']

pyplot.pie(product_sales,labels=items,explode=(0,0,0,0.1,0,0),autopct='%1.1f%%',startangle=90,shadow=True)
pyplot.show()
```
