```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['What year were you in last year (2023) ?'] = le.fit_transform(df['What year were you in last year (2023) ?'])
df['What faculty does your degree fall under?'] = le.fit_transform(df['What faculty does your degree fall under?'])
df['Your Accommodation Status Last Year (2023)'] = le.fit_transform(df['Your Accommodation Status Last Year (2023)'])
df['Monthly Allowance in 2023'] = le.fit_transform(df['Monthly Allowance in 2023'])
df['Were you on scholarship/bursary in 2023?'] = le.fit_transform(df['Were you on scholarship/bursary in 2023?'])
df['Additional amount of studying (in hrs) per week'] = le.fit_transform(df['Additional amount of studying (in hrs) per week'])
df['How often do you go out partying/socialising during the week? '] = le.fit_transform(df['How often do you go out partying/socialising during the week? '])
df['On a night out, how many alcoholic drinks do you consume?'] = le.fit_transform(df['On a night out, how many alcoholic drinks do you consume?'])
df['How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)'] = le.fit_transform(df['How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)'])
df['How many modules have you failed thus far into your studies?'] = le.fit_transform(df['How many modules have you failed thus far into your studies?'])
df['Are you currently in a romantic relationship?'] = le.fit_transform(df['Are you currently in a romantic relationship?'])
df['Do your parents approve alcohol consumption?'] = le.fit_transform(df['Do your parents approve alcohol consumption?'])
df['How strong is your relationship with your parent/s?'] = le.fit_transform(df['How strong is your relationship with your parent/s?'])
```
